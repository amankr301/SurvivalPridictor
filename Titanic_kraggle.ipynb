```python
!pip install xgboost
```

```python
#XGBOOST
from xgboost import XGBClassifier
```

```python
model = XGBClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
#CONFUSION MATRIX
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm 
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator= model, X = X_train, y = y_train, cv = 10)
print(accuracies.mean(), accuracies.std())
```

```python
accuracies.mean()
```

```python
#HYPERTUNNING
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'learning_rate' : [0.05, 0.75, 0.1, 0.2, 0.3, 0.4]}
```

```python
grid = GridSearchCV(estimator = model ,param_grid =  parameters, verbose = 10)
grid.fit(X_train,y_train)
```

```python
print(grid.best_score_)
```

```python
#GAMMA
```

```python
parameter = { 'gamma' : [0, 0.05, 0.1]}
```

```python
#MAX_Depth
```

```python
parameter = {'max_depth':[2,3,4,5,6]}
```

```python
# n_estimators
```

```python
#min_child_weight
```

```python
parameter = {'min_child_weight': [1,2,3,4,5]}
```

```python
model = XGBClassifier(learning_rate = 0.05, gamma = 0,
                     max_depth = 4, n_estimators = 150, random_state = 42)
```

```python
#FINAL MODEL
```

```python
model = XGBClassifier(learning_rate = 0.05, gamma = 0, max_depth = 4, 
                      n_estimators = 150,min_child_weight= 2, random_state = 42)
```

```python
grid = GridSearchCV(estimator=model, param_grid=parameter, verbose = 10)
grid.fit(X_train,y_train)
```
